# ISO/IEC 42001 Compliance Automation Demo — NovaPharma AI

## Scenario
NovaPharma AI is an 800-employee pharmaceutical company developing AI systems for drug discovery and clinical trials:
- **DrugDiscoveryGPT**: Generative AI analyzing molecular compounds and identifying promising drug candidates
- **ClinicalTrialOptimizer**: ML system optimizing patient cohort selection and trial protocol parameters

**Current State**: NovaPharma has made significant progress on AI governance compared to DataVault, with:
- AI governance charter already published (v2.1)
- Formal risk assessment methodology in draft form
- Ad-hoc monitoring practices for accuracy and bias
- Partial competence framework in place
- No formal impact assessment process yet

**Regulatory Context**: Operating in highly regulated pharmaceutical industry (FDA oversight, clinical trial requirements), making ISO 42001 certification both strategic and compliance-aligned.

**Goal**: Complete ISO/IEC 42001 certification within 9 months (by Q4 2026).

## ISO/IEC 42001 Quick Reference

| Clause | Area | Key Requirement |
|--------|------|----------------|
| 4 | Context | Understand organization context and stakeholder needs for AI |
| 5 | Leadership & Policy | Leadership commitment, AI policy, roles and responsibilities |
| 6 | Planning | Risk assessment, AI objectives, change management |
| 7 | Support | Resources, competence, awareness, documentation |
| 8 | Operations | AI lifecycle management, impact assessment, third-party governance |
| 9 | Performance | Monitoring, measurement, internal audit |
| 10 | Improvement | Corrective actions, continual improvement |

## Readiness Scoring Scale

| Score | Level | Description |
|-------|-------|-------------|
| 1 | Not Started | No controls or processes in place |
| 2 | Initiated | Some awareness; initial steps taken |
| 3 | In Progress | Partial implementation; inconsistent application |
| 4 | Implemented | Controls in place and operating consistently |
| 5 | Optimized | Continuously improved; benchmarked against best practices |

**Certification Threshold**: Average readiness score ≥ 4.0 (80%+) across all control areas.

## Objective
Demonstrate how a more mature organization (NovaPharma) conducts a compliance assessment and chart a faster path to ISO 42001 certification by building on existing governance foundations.

## 2a. Load Controls from Excel

```python
# Load controls from Excel file and create novapharma_controls dictionary
# This synchronizes the notebook with the Excel assessment data

excel_file = 'iso42001_demo.xlsx'
excel_loaded = False

try:
    # Read the Gap Analysis Checklist sheet from Excel
    df_excel = pd.read_excel(excel_file, sheet_name='Gap Analysis Checklist')
    
    # Reconstruct novapharma_controls from Excel data
    # Group by clause for the nested dictionary structure
    novapharma_controls = {}
    
    for _, row in df_excel.iterrows():
        control_id = row['Control ID']
        clause_num = row['ISO 42001 Clause']
        control_area = row['Control Area']
        clause_key = f'Clause {clause_num}: {control_area}'
        
        if clause_key not in novapharma_controls:
            novapharma_controls[clause_key] = []
        
        # Extract readiness score from Gap Description (which contains the score)
        readiness_score = row['Gap Description']
        if pd.isna(readiness_score):
            readiness_score = 3  # Default to In Progress if missing
        else:
            readiness_score = int(readiness_score)
        
        control_dict = {
            'control_id': control_id,
            'description': row['Requirement Description'],
            'current_state': row['Current State'],
            'evidence': row['Evidence Available'],
            'readiness_score': readiness_score
        }
        
        novapharma_controls[clause_key].append(control_dict)
    
    excel_loaded = True
    print('✓ Controls and scores loaded from Excel file: iso42001_demo.xlsx')
    print(f'✓ Total controls: {len(df_excel)}')
    
except Exception as e:
    print(f'✗ Error loading from Excel: {e}')
    print('✗ Will use hardcoded fallback instead.')
    excel_loaded = False
```

## 3. Build Gap Analysis DataFrame

```python
# Convert control requirements to DataFrame for analysis

data = []
for clause, controls in novapharma_controls.items():
    for control in controls:
        data.append({
            'clause': clause,
            'control_id': control['control_id'],
            'description': control['description'],
            'current_state': control['current_state'],
            'evidence': control['evidence'],
            'readiness_score': control['readiness_score']
        })

df_assessment = pd.DataFrame(data)

print('\n' + '='*100)
print('NOVAPHARMA AI — ISO 42001 COMPLIANCE ASSESSMENT')
print('='*100)
print(f'\nTotal Controls Assessed: {len(df_assessment)}')
print(f'Overall Average Readiness Score: {df_assessment["readiness_score"].mean():.2f}/5.0')
print(f'Overall Readiness Percentage: {(df_assessment["readiness_score"].mean() / 5.0 * 100):.1f}%')
print(f'Certification Target (80%): Gap = {80 - (df_assessment["readiness_score"].mean() / 5.0 * 100):.1f} percentage points')

print('\n' + '-'*100)
print('\nDetailed Gap Analysis by Control:')
print('-'*100)
for _, row in df_assessment.iterrows():
    status = 'COMPLETE' if row['readiness_score'] >= 4 else ('AT RISK' if row['readiness_score'] >= 3 else 'CRITICAL')
    print(f"\n[{row['control_id']}] {row['description']}")
    print(f"  Clause: {row['clause']}")
    print(f"  Current State: {row['current_state']} | Readiness: {row['readiness_score']}/5.0 [{status}]")
    print(f"  Evidence: {row['evidence']}")

print('\n' + '-'*100)
```

## 4. Calculate Readiness by Control Area

```python
# Calculate per-area readiness scores

readiness_summary = []
for clause in novapharma_controls.keys():
    controls = novapharma_controls[clause]
    scores = [c['readiness_score'] for c in controls]
    avg_score = sum(scores) / len(scores)
    pct = (avg_score / 5.0) * 100
    status = 'Complete' if pct >= 80 else ('In Progress' if pct >= 60 else 'Not Started')
    
    readiness_summary.append({
        'Clause': clause,
        'Controls': len(scores),
        'Avg_Score': round(avg_score, 2),
        'Readiness_%': round(pct, 1),
        'Status': status
    })

df_readiness = pd.DataFrame(readiness_summary)

print('\n' + '='*100)
print('READINESS SUMMARY BY CONTROL AREA')
print('='*100)
print(df_readiness.to_string(index=False))

print('\n\nOVERALL ASSESSMENT:')
overall_score = df_assessment['readiness_score'].mean()
overall_pct = (overall_score / 5.0) * 100
print(f'  Overall Score: {overall_score:.2f}/5.0')
print(f'  Overall Readiness: {overall_pct:.1f}%')
print(f'  Certification Status: {"Ready for Certification" if overall_pct >= 80 else ("On Track" if overall_pct >= 60 else "Needs Work")}')
print(f'  Estimated Time to Certification: 6-9 months (from today)')
print('\n' + '='*100)
```

## 7. Visualization: Readiness Charts

```python
# Create readiness radar chart and gap bar chart for NovaPharma

# Prepare data for radar chart (map by control area from clause)
clause_mapping = {}
for clause, controls in novapharma_controls.items():
    clause_short = clause.split(':')[0].replace('Clause ', 'C')
    clause_mapping[clause_short] = [c['readiness_score'] for c in controls]

areas_for_radar = list(clause_mapping.keys())
scores_for_radar = [np.mean(clause_mapping[area]) for area in areas_for_radar]

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# ===== RADAR CHART =====
num_areas = len(areas_for_radar)
angles = np.linspace(0, 2 * np.pi, num_areas, endpoint=False).tolist()
scores_for_radar_plot = scores_for_radar + [scores_for_radar[0]]
angles += [angles[0]]

# Create radar chart
ax1 = plt.subplot(121, projection='polar')
ax1.plot(angles, scores_for_radar_plot, 'o-', linewidth=2.5, color='#2E86AB', label='Current Readiness')
ax1.fill(angles, scores_for_radar_plot, alpha=0.25, color='#2E86AB')

# Add 80% certification threshold
threshold_line = [4.0] * (num_areas + 1)
ax1.plot(angles, threshold_line, '--', linewidth=2.5, color='#A23B72', label='Certification Threshold (80%)')
ax1.fill(angles, threshold_line, alpha=0.1, color='#A23B72')

# Customize radar chart
ax1.set_xticks(angles[:-1])
ax1.set_xticklabels(areas_for_radar, size=11, weight='bold')
ax1.set_ylim(0, 5)
ax1.set_yticks([1, 2, 3, 4, 5])
ax1.set_yticklabels(['1', '2', '3', '4', '5'], size=9)
ax1.set_rlabel_position(0)
ax1.grid(True, linewidth=0.8)
ax1.legend(loc='upper right', bbox_to_anchor=(1.25, 1.1), fontsize=10)
ax1.set_title('NovaPharma AI — Readiness by Control Area\n(Radar Chart)', size=13, weight='bold', pad=20)

# ===== GAP BAR CHART =====
# Prepare data: sort controls by readiness score
df_sorted = df_assessment.sort_values('readiness_score', ascending=True)
controls = df_sorted['control_id'].tolist()
scores = df_sorted['readiness_score'].tolist()

# Define color coding based on gap severity
colors = []
for score in scores:
    if score < 2:
        colors.append('#D62828')  # Red - Critical
    elif score < 3:
        colors.append('#F77F00')  # Orange - At Risk
    elif score < 4:
        colors.append('#FCBF49')  # Yellow - In Progress
    else:
        colors.append('#06A77D')  # Green - Implemented

# Create horizontal bar chart
y_pos = np.arange(len(controls))
ax2.barh(y_pos, scores, color=colors, edgecolor='black', linewidth=0.5)

# Add certification threshold line
ax2.axvline(x=4.0, color='#A23B72', linestyle='--', linewidth=2.5, label='Certification Threshold (4.0)')

# Customize bar chart
ax2.set_yticks(y_pos)
ax2.set_yticklabels(controls, size=10)
ax2.set_xlabel('Readiness Score (0-5)', size=12, weight='bold')
ax2.set_xlim(0, 5)
ax2.set_xticks([0, 1, 2, 3, 4, 5])
ax2.grid(axis='x', alpha=0.3, linewidth=0.8)
ax2.set_title('Control Readiness Gap Analysis\n(Color-coded by Severity)', size=13, weight='bold')

# Add value labels on bars
for i, (control, score) in enumerate(zip(controls, scores)):
    ax2.text(score + 0.1, i, f'{score:.0f}', va='center', fontsize=9, weight='bold')

# Add legend for color coding
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='#D62828', edgecolor='black', label='Critical Gap (< 2)'),
    Patch(facecolor='#F77F00', edgecolor='black', label='At Risk (2-3)'),
    Patch(facecolor='#FCBF49', edgecolor='black', label='In Progress (3-4)'),
    Patch(facecolor='#06A77D', edgecolor='black', label='Implemented (≥ 4)')
]
ax2.legend(handles=legend_elements, loc='lower left', fontsize=10)

plt.tight_layout()
plt.savefig('novapharma_readiness_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print('\n' + '='*100)
print('VISUALIZATION ANALYSIS')
print('='*100)
print('\nRadar Chart Insights:')
print(f'  • Shows readiness across {num_areas} control areas (Clauses 5, 6, 7, 8, 9)')
print(f'  • Purple dashed line represents 80% certification threshold (4.0/5.0)')
print(f'  • All areas must reach the threshold to achieve certification')
print(f'  • Current state: {sum(1 for s in scores_for_radar if s >= 4.0)}/{num_areas} areas at or above threshold')

print('\nGap Bar Chart Insights:')
critical = sum(1 for s in scores if s < 2)
at_risk = sum(1 for s in scores if 2 <= s < 3)
in_prog = sum(1 for s in scores if 3 <= s < 4)
implemented = sum(1 for s in scores if s >= 4)
print(f'  • Critical gaps (red, < 2): {critical} controls')
print(f'  • At risk (orange, 2-3): {at_risk} controls')
print(f'  • In progress (yellow, 3-4): {in_prog} controls')
print(f'  • Implemented (green, ≥ 4): {implemented} controls')

print('\nCharts saved as: novapharma_readiness_analysis.png')
print('='*100)
```
