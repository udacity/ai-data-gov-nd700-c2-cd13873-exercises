## Step 1: Define Risk TiersThe EU AI Act defines four risk levels. Let's encode the key classification rules.

```python
# EU AI Act risk tier definitions — these are the classification rulesRISK_TIERS = {    "Unacceptable": {        "description": "Prohibited AI practices",        "triggers": ["social_scoring", "real_time_biometric_public", "manipulation_vulnerable"],        "action": "BANNED — cannot be deployed in the EU"    },    "High-Risk": {        "description": "Significant impact on fundamental rights or safety",        "triggers": ["employment_decisions", "biometric_identification", "critical_infrastructure",                     "law_enforcement", "education_access", "credit_scoring", "worker_monitoring"],        "action": "Requires conformity assessment, documentation, human oversight, monitoring"    },    "Limited": {        "description": "Transparency obligations apply",        "triggers": ["chatbot_interaction", "emotion_recognition", "deepfake_generation"],        "action": "Must disclose AI usage to users"    },    "Minimal": {        "description": "No specific EU AI Act requirements",        "triggers": ["spam_filter", "recommendation_basic", "optimization_no_rights_impact"],        "action": "Voluntary codes of conduct"    }}for tier, info in RISK_TIERS.items():    print(f"\n{tier}: {info['description']}")    print(f"  Action: {info['action']}")
```

## Step 2: Build the ClassifierThe instructor builds a rule-based function that takes system characteristics and returns a risk tier classification with reasoning.

```python
def classify_eu_ai_act(system_name, characteristics):    """    Classify an AI system's risk tier under the EU AI Act.        Parameters:        system_name (str): Name of the AI system        characteristics (dict): System attributes including:            - sector: str (e.g., 'logistics', 'hr', 'surveillance')            - uses_biometric: bool            - affects_employment: bool            - monitors_workers: bool              - impacts_fundamental_rights: bool            - is_safety_critical: bool            - social_scoring: bool            - interacts_with_users: bool        Returns:        dict with tier, reasoning, required_actions, applicable_articles    """    reasons = []    articles = []        # Check Unacceptable first (prohibited practices)    if characteristics.get("social_scoring", False):        return {            "system": system_name,            "tier": "Unacceptable",            "reasoning": ["System performs social scoring — prohibited under Article 5"],            "required_actions": ["MUST NOT BE DEPLOYED. Remove immediately."],            "applicable_articles": ["Article 5"]        }        # Check High-Risk (Annex III categories)    if characteristics.get("affects_employment", False):        reasons.append("Makes or influences employment decisions (Annex III, area 4)")        articles.extend(["Article 6", "Annex III(4)"])        if characteristics.get("monitors_workers", False):        reasons.append("Monitors workers in the workplace (Annex III, area 4)")        articles.extend(["Article 6", "Annex III(4)"])        if characteristics.get("uses_biometric", False):        reasons.append("Uses biometric data processing (Annex III, area 1)")        articles.extend(["Article 6", "Annex III(1)"])        if characteristics.get("is_safety_critical", False):        reasons.append("Safety-critical system (Annex III, area 2)")        articles.extend(["Article 6", "Annex III(2)"])        if characteristics.get("impacts_fundamental_rights", False):        reasons.append("Impacts fundamental rights of individuals")        articles.append("Article 6")        if reasons:        return {            "system": system_name,            "tier": "High-Risk",            "reasoning": reasons,            "required_actions": [                "Conduct conformity assessment (Article 43)",                "Implement risk management system (Article 9)",                "Ensure data governance (Article 10)",                "Maintain technical documentation (Article 11)",                "Enable human oversight (Article 14)",                "Ensure accuracy and robustness (Article 15)",                "Register in EU database (Article 49)"            ],            "applicable_articles": list(set(articles))        }        # Check Limited Risk    if characteristics.get("interacts_with_users", False):        return {            "system": system_name,            "tier": "Limited",            "reasoning": ["System interacts directly with users — transparency obligation applies"],            "required_actions": ["Disclose AI involvement to users (Article 52)"],            "applicable_articles": ["Article 52"]        }        # Default: Minimal    return {        "system": system_name,        "tier": "Minimal",        "reasoning": ["No high-risk triggers identified"],        "required_actions": ["No mandatory requirements. Consider voluntary code of conduct."],        "applicable_articles": []    }
```

## Step 3: Define Our AI SystemsNow let's describe each of AutoPilot Logistics' three AI systems using the characteristics our classifier expects.

```python
# Define AutoPilot Logistics AI systems with their characteristicsautopilot_systems = {    "RouteGenius": {        "description": "GenAI route optimization — plans delivery routes, predicts traffic",        "sector": "logistics",        "uses_biometric": False,        "affects_employment": False,        "monitors_workers": False,        "impacts_fundamental_rights": False,        "is_safety_critical": False,        "social_scoring": False,        "interacts_with_users": False  # Drivers use it but it's a tool, not a chatbot    },    "WarehouseWatch": {        "description": "AI worker safety monitoring — camera-based unsafe behavior detection",        "sector": "workplace_monitoring",        "uses_biometric": True,   # Uses camera/video of workers        "affects_employment": False,  # Doesn't directly make employment decisions        "monitors_workers": True,  # Core function is monitoring workers        "impacts_fundamental_rights": True,  # Worker privacy and dignity        "is_safety_critical": True,  # Safety system        "social_scoring": False,        "interacts_with_users": False    },    "HireBot": {        "description": "AI applicant screening — ranks job applicants for warehouse positions",        "sector": "hr",        "uses_biometric": False,        "affects_employment": True,  # Directly screens job applicants        "monitors_workers": False,        "impacts_fundamental_rights": True,  # Employment is a fundamental right        "is_safety_critical": False,        "social_scoring": False,        "interacts_with_users": True  # Applicants interact with the screening system    }}print(f"Defined {len(autopilot_systems)} AI systems for classification")for name, chars in autopilot_systems.items():    print(f"  - {name}: {chars['description']}")
```

## Step 4: Run the Classifier

```python
# Classify each systemresults = {}for name, chars in autopilot_systems.items():    results[name] = classify_eu_ai_act(name, chars)# Display resultsfor name, result in results.items():    print(f"\n{'='*60}")    print(f"System: {name}")    print(f"Risk Tier: {result['tier']}")    print(f"\nReasoning:")    for r in result['reasoning']:        print(f"  - {r}")    print(f"\nRequired Actions:")    for a in result['required_actions']:        print(f"  - {a}")    if result['applicable_articles']:        print(f"\nApplicable Articles: {', '.join(result['applicable_articles'])}")
```
